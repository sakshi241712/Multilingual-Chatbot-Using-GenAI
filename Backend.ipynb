```python
from indic_transliteration import sanscript
from indic_transliteration.sanscript import SchemeMap, SCHEMES, transliterate
from googletrans import Translator

# Initialize the Translator
translator = Translator()

def process_hindi_bot(user_roman_input):
    # STEP 1: Transliterate (English letters -> Hindi Script)
    # Example: "namaste" -> "नमस्ते"
    hindi_script = transliterate(user_roman_input, sanscript.ITRANS, sanscript.DEVANAGARI)
    print(f"Detected Hindi Script: {hindi_script}")

    # STEP 2: Translate to English for the LLM
    translation_to_en = translator.translate(hindi_script, src='hi', dest='en')
    english_query = translation_to_en.text
    print(f"English translation for LLM: {english_query}")

    # STEP 3: LLM Reasoning (Simulated here)
    # In a real app, you'd send 'english_query' to Gemini or GPT-4
    llm_english_response = "Hello! I am doing well, how can I help you today?"

    # STEP 4: Translate LLM Response back to Hindi
    translation_to_hi = translator.translate(llm_english_response, src='en', dest='hi')
    final_hindi_output = translation_to_hi.text

    return final_hindi_output

# --- TEST ---
user_input = "sheegra bahar aaiye , anyatha mujhe maarg par aanaa hoga. ky aap sajjah he?" # User types in Roman script
response = process_hindi_bot(user_input)
print(f"Bot Response: {response}")
```
